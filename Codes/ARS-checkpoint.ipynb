## Nettoyage

```python
# Liste des termes à remplacer
alphabet_alias = [ "Google", "Waymo","GOOGLE", "google","Google DeepMind", "Google DeepMind (A)", "deepmind", "Google Deep Ming", "Google Cloud" ]
openai_alias =["Open AI", "Open Ai", "OpenAi"]
meta_alias=["meta", "Facebook", "Meta"]
amazon_alias =["AWS"]
YT_alias =["Youtube"]
# Fonction de nettoyage pour supprimer les suffixes entre parenthèses
def remove_parentheses(text):
    return re.sub(r"\s*\([^)]*\)$", "", text) 
    
# Application sur les deux colonnes
for col in ['Entreprise A', 'Entreprise B']:
    deal[col] = deal[col].astype(str).str.strip()            # trim espaces
    deal[col] = deal[col].apply(remove_parentheses)          # enlève parenthèse finale
    deal[col] = deal[col].replace(alphabet_alias, "Alphabet", regex=False)  # remplace variantes Alphabet
    deal[col] = deal[col].replace(openai_alias, "OpenAI", regex=False)  # remplace variantes Alphabet
    deal[col] = deal[col].replace(meta_alias, "Meta", regex=False)  # remplace variantes Alphabet
    deal[col] = deal[col].replace(amazon_alias, "Amazon", regex=False)  # remplace variantes Alphabet
    deal[col] = deal[col].replace(YT_alias, "YouTube", regex=False)  # remplace variantes Alphabet
    deal[col] = deal[col].str.strip()   


display((deal["Entreprise A"].unique()))
display((deal["Entreprise B"].unique()))
```

## Graphe non orienté complet

```python
# Créer le graphe non orienté
G = nx.from_pandas_edgelist(deal, 'Entreprise A', 'Entreprise B')
print(f"Nombre d'entreprises : {G.number_of_nodes()} | Nombre d'alliances : {G.number_of_edges()}")
```

```python
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.4, seed=42)  # positionnement esthétique
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='skyblue', alpha=0.8)
nx.draw_networkx_edges(G, pos, alpha=0.3)
nx.draw_networkx_labels(G, pos, font_size=8)

plt.title("Réseau d'alliances interentreprises", fontsize=14)
plt.axis('off')
plt.show()
```

```python
from pyvis.network import Network

# Créer un réseau interactif
net = Network(height="750px", width="100%", bgcolor="#ffffff", font_color="black")

# Charger les données du graphe NetworkX
net.from_nx(G)

# Options esthétiques
net.force_atlas_2based()  # disposition plus harmonieuse
net.show_buttons(filter_=['physics'])  # pour ajuster l'espacement dans le navigateur
net.write_html("reseau_alliances.html")
```

```python
# =========================
# Multiplex PyVis (vraies couches)
# =========================
import math
import pandas as pd

# --- 0) Paramètres ---
COUPLING_INTERCOUCHES = True     # relier les copies d'une même entreprise entre couches
ESPACE_VERTICAL = 300            # distance verticale entre couches (px)
TAILLE_NOEUD = 20                # rayon visuel des noeuds
PHYSICS = False                  # False = positions fixes (recommandé pour un multiplex)
HTML_OUT = "multiplex_layers.html"

# --- 1) Nettoyage des données ---
# On suppose que 'deal' est déjà un DataFrame en mémoire,
# sinon : deal = pd.read_csv("ton_fichier.csv")
deal2 = deal.dropna(subset=["Entreprise A", "Entreprise B", "Type d'alliance"]).copy()

# Normaliser les types (facultatif mais utile)
deal2["Type d'alliance"] = deal2["Type d'alliance"].astype(str).str.strip()

layers = sorted(deal2["Type d'alliance"].unique())
if not layers:
    raise ValueError("Aucun 'Type d'alliance' valide trouvé (colonne vide ou NaN).")

# --- 2) Construire le graphe multiplex (nœuds = (Entreprise, Couche)) ---
# Un graphe “logique” pour calculer les positions et tenir les attributs
MG = nx.Graph()

# Arêtes intra-couche
for L in layers:
    sub = deal2[deal2["Type d'alliance"] == L]
    for _, r in sub.iterrows():
        u = (r["Entreprise A"], L)
        v = (r["Entreprise B"], L)
        MG.add_node(u, label=r["Entreprise A"], layer=L)
        MG.add_node(v, label=r["Entreprise B"], layer=L)
        MG.add_edge(u, v, layer=L, intralayer=True)

# (Optionnel) Couplage inter-couches : relier les copies d'une même entreprise entre couches
if COUPLING_INTERCOUCHES:
    entreprises = pd.unique(pd.concat([deal2["Entreprise A"], deal2["Entreprise B"]], ignore_index=True))
    for n in entreprises:
        copies = [(n, L) for L in layers if (n, L) in MG]
        # trier par ordre de layers pour relier dans l'ordre
        copies = sorted(copies, key=lambda x: layers.index(x[1]))
        for i in range(len(copies) - 1):
            MG.add_edge(copies[i], copies[i+1], intralayer=False, weight=0.1)

# --- 3) Calcul des positions (x,y) par couche ---
# On calcule une disposition par couche, puis on fige le y selon la couche.
pos = {}
for idx, L in enumerate(layers):
    nodes_L = [n for n in MG if isinstance(n, tuple) and n[1] == L]
    H = MG.subgraph(nodes_L).copy()
    if len(H) == 0:
        continue

    # Layout local (spring ou circular)
    # circular_layout = plus stable et “propre” pour couches petites
    pos_local = nx.circular_layout(H, scale=200.0)  # x,y typiquement ~[-200, 200]

    # Appliquer un y constant par couche
    y_layer = -idx * ESPACE_VERTICAL
    for n, (x, y) in pos_local.items():
        pos[n] = (float(x), float(y_layer))

# --- 4) PyVis : création + ajout des nœuds/arrêtes avec positions fixes ---
net = Network(height="800px", width="100%", bgcolor="#ffffff", font_color="black", notebook=False)
if not PHYSICS:
    # désactive globalement la physique pour respecter les x/y donnés
    net.options.physics.enabled = False

# Palette simple par couche (PyVis colorie par "group")
palette = ["#1f77b4","#2ca02c","#d62728","#9467bd","#8c564b","#e377c2","#7f7f7f","#bcbd22","#17becf"]
layer_color = {L: palette[i % len(palette)] for i, L in enumerate(layers)}

# Helper : ID PyVis doit être sérialisable → utiliser une string unique
def node_id(name, layer):
    return f"{name}||{layer}"

# Ajouter nœuds
for n, data in MG.nodes(data=True):
    name, layer = n
    x, y = pos.get(n, (0.0, 0.0))
    net.add_node(
        node_id(name, layer),
        label=f"{name} ({layer})",               # label affiché
        title=f"{name}<br>Couche: {layer}",      # tooltip
        x=x, y=y,
        physics=PHYSICS,                         # False = position figée
        size=TAILLE_NOEUD,
        color=layer_color[layer],
        group=layer                              # pour la légende par groupe
    )

# Ajouter arêtes (intra = pleines, inter = pointillées)
for u, v, ed in MG.edges(data=True):
    u_name, u_layer = u
    v_name, v_layer = v
    intra = ed.get("intralayer", False)
    if intra:
        color = "#444444"
        dashes = False
        width = 1.5
        title = f"Type: {u_layer}"
    else:
        color = "#aaaaaa"
        dashes = True
        width = 1
        title = "Couplage inter-couches"

    net.add_edge(
        node_id(u_name, u_layer),
        node_id(v_name, v_layer),
        color=color,
        width=width,
        dashes=dashes,
        title=title
    )

# Boutons de physique/interaction (optionnel)
net.show_buttons(filter_=["physics"])

# Export HTML (ouvre aussi le navigateur)
net.show(HTML_OUT)
print(f"✅ Multiplex exporté dans: {HTML_OUT}")
```
